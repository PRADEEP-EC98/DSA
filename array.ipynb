# FIND THE LARGEST ELEMENT IN THE ARRAY

```python
def largest(arr):
    if not arr:
        return None
    largest = float('-inf')
    for num in arr:
        if num>largest:
            largest = num
    return largest

arr = [1,55,89,77,100,67]
print(f'Largest Element= {largest(arr)}')
```

# FIND THE SECOND LARGEST ELEMENT IN THE ARRAY

```python
def second_largest(arr):
    if len(arr) < 2:
        return None

    largest = float("-inf")
    second_largest = float('-inf')

    for num in arr:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num
    return second_largest

arr = [1,55,89,77,100,67]
print(f'Second Largest Element= {second_largest(arr)}')
```

# REVERSE ARRAY

```python
def reverse_array(arr):
    if not arr:
        return None

    start = 0
    end = len(arr)-1

    while start < end:                                 
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1
    return arr

arr = [1,55,89,77,100,67]
print(f'Reversed Array = {reverse_array(arr)}')
```

# CHECK IS ARRAY SORTED

```python
def is_sorted(arr):
    if not arr:
        return None
    if len(arr) == 1:
        return True

    for i in range(1, len(arr)):
        if arr[i] < arr[i-1]:
            return False

    return True
    
arr = [1,55,89,77,100,67]
print(f'Is array sorted = {is_sorted(arr)}')
arr = [1,2,3,4,5,6,7,8]
print(f'Is array sorted = {is_sorted(arr)}')

    
```

# Remove duplicates from sorted array

```python
def duplicates(arr):
    if not arr:
        return None

    i = 0
    for j in range(1, len(arr)):
        if arr[j] != arr[i]:
            i+= 1
            arr[i] = arr[j]
            
    return i+1

arr = [1,1,2,2,2,3,3,4,4,4,4,5,5,6,7,7,8]
new_len = duplicates(arr)
print(f'New Length = {new_len}')
print(f"Array after removing duplicates {arr[:new_len]}")
```

# MOVE ALL ZEROS TO THE END

```python
def move_zero(arr):
    if not arr:
        return None

    i = 0

    for j in range(len(arr)):
        if arr[j] != 0:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    return arr
arr = [0,1,0,2,3,4,5,0,6]
print(f"Result = {move_zero(arr)}")
arr = []
print(move_zero(arr))
```

```python
def move_zeros(arr):
    pos = 0  
    
    for num in arr:
        if num != 0:
            arr[pos] = num
            pos += 1
            
    while pos < len(arr):
        arr[pos] = 0
        pos += 1
    
    return arr

arr = [0, 1, 0, 3, 12]
print("Result:", move_zeros(arr))
```

# LEFT ROTATE OF ARRAY BY ONE

```python
def left_rotate(arr):
    if len(arr) <= 1:
        return arr

    first = arr[0]
    for i in range(1,len(arr)):
        arr[i-1] = arr[i]

    arr[-1] = first

    return arr

arr = [1,2,3,4,5]
print(f"Result = {left_rotate(arr)}")

    
```

# FINDING MISSING VALUE

```python
def find(arr):
    if not arr:
        return None

    n = len(arr)

    expected_sum = (n*(n+1))//2
    actual_sum = sum(arr)

    return expected_sum - actual_sum

# Array of size 4 containing numbers from 0 to 4
arr = [0,1,3,4]
print(f"Missing Value = {find(arr)}")

```
